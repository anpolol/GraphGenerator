```python
from scipy.stats import betabinom, beta
import matplotlib.pyplot as plt
import numpy as np
```

```python
rv = betabinom(10,2.5,2.5)
x = np.arange(rv.ppf(0.01),
              rv.ppf(0.99))
fig, ax = plt.subplots(1, 1)
ax.plot(x, rv.pmf(x), 'bo', ms=8, label='betabinom pmf')
ax.vlines(x, 0, rv.pmf(x), colors='b', lw=5, alpha=0.5)
plt.show()
```

```python
rv = beta(7.5, 7.5)
plt.hist(rv.rvs(500*500), bins=40,density=True,alpha=0.5, label = 'Retrieved')
plt.hist(beta(1,4).rvs(500*500),bins=40,density=True,alpha=0.5,label = 'Desired')
plt.legend()
plt.show()
```
