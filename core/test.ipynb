```python
from scipy.stats import betabinom
import matplotlib.pyplot as plt
import numpy as np
```

```python
rv = betabinom(10,5.5,1.5)
x = np.arange(rv.ppf(0.01),
              rv.ppf(0.99))
fig, ax = plt.subplots(1, 1)
ax.plot(x, rv.pmf(x), 'bo', ms=8, label='betabinom pmf')
ax.vlines(x, 0, rv.pmf(x), colors='b', lw=5, alpha=0.5)
plt.show()
```
